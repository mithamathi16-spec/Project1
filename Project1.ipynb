```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Department': ['Sales', 'HR', 'IT', 'Sales', 'IT', 'HR', 'Sales', 'IT', 'HR', 'Sales'],
    'MonthlyIncome': [25000, 30000, 50000, 22000, 55000, 28000, 24000, 60000, 27000, 23000],
    'YearsAtCompany': [1, 5, 7, 2, 8, 4, 1, 9, 3, 2],
    'Attrition': ['Yes', 'No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'Yes']
}

df = pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.show()
```

```python
sns.countplot(x='Department', hue='Attrition', data=df)
plt.show()
```

```python
df['Attrition'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Employee Attrition")
plt.show()
```

```python
# Dashboard Layout
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# 1. Attrition Pie Chart
df['Attrition'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    ax=axes[0,0]
)
axes[0,0].set_title("Employee Attrition")

# 2. Department-wise Attrition
sns.countplot(x='Department', hue='Attrition',
              data=df, ax=axes[0,1])
axes[0,1].set_title("Department-wise Attrition")

# 3. Income Distribution
sns.boxplot(x='Attrition', y='MonthlyIncome',
            data=df, ax=axes[1,0])
axes[1,0].set_title("Income vs Attrition")

# 4. Experience Distribution
sns.barplot(x='Attrition', y='YearsAtCompany',
            data=df, ax=axes[1,1])
axes[1,1].set_title("Experience vs Attrition")

plt.tight_layout()
plt.show()
```
